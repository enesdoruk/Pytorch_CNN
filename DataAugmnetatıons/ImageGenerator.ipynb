```python
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
import os
```

```python
datagen = ImageDataGenerator(rotation_range=10,
                            width_shift_range=0.1,
                            height_shift_range=0.1,
                            shear_range=0.2,
                            zoom_range=0.3,
                            brightness_range=(0.4,0.7),
                            horizontal_flip=True,
                            vertical_flip=False,
                            fill_mode='nearest')
data0 = os.listdir("C:/Users/ENES DORUK/Desktop/CNN_Harf/veriler/training_set/0/")
for j in range(len(data0)):
    img=load_img("C:/Users/ENES DORUK/Desktop/CNN_Harf/veriler/training_set/0/{}".format(data0[j]))
    x = img_to_array(img)
    x = x.reshape((1,) + x.shape)
    
    i=0
    
    for batch in datagen.flow(x, batch_size=1, save_to_dir="C:/Users/ENES DORUK/Desktop/CNN_Harf/veriler/test_set/0/",
                             save_format="png"):
        i +=1
        if i == 5:
            break
```
